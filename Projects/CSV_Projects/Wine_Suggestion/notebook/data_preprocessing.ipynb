# Create Documents with CSV data.

```python
from langchain.schema import Document
from tqdm import tqdm
```

```python
def create_document_from_tuple(t):
  return Document(
      page_content=t.description,
      metadata={
            'country': t.country,
            'province': t.province,
            'name': t.title,
            'variety':t.variety,
            'winery':t.winery
        }
  )
```

```python
docs = []

for row in tqdm(df.itertuples(index = False)):
  documents = create_document_from_tuple(row)
  docs.append(documents)
```
